# 2. Model Implementation

```python
#●	Construct a basic ANN model using your chosen high-level neural network library. Ensure your model includes at least one hidden layer.
#●	Divide the dataset into training and test sets.
#●	Train your model on the training set and then use it to make predictions on the test set.
# X & Y
X = df.drop(columns=['letter'])
y = df['letter']

# Splitting data into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_normalization , label_encoded, test_size=0.2, random_state=42)
```

```python
# !pip install tensorflow

# Import necessary libraries
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split

# Define the model architecture
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(X_normalization.shape[1],)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(26, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_normalization, label_encoded, test_size=0.2, random_state=42)

# Train the model
model.fit(X_train, y_train, epochs=10)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(X_test, y_test)
print('Test loss:', test_loss)
print('Test accuracy:', test_acc)

# Print the classification report

# Make predictions on the test set and convert to discrete class labels
predictions = model.predict(X_test)
predicted_classes = np.argmax(predictions, axis=1)

# Generate a classification report
from sklearn.metrics import classification_report
report = classification_report(y_test, predicted_classes,target_names=label_encoder.classes_)
print("Classification Report:")
print(report)
```

# 3. Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
# Defining  the parameter grid here

param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'sgd'],
    'learning_rate_init': [0.001, 0.01, 0.1]
}
# Creating the ANN classifier
ann_classifier = MLPClassifier() # Create an instance of the classifier

# Creating  a grid search object
grid_search = GridSearchCV(ann_classifier, param_grid, cv=3, scoring='accuracy')
grid_search.fit(X_train, y_train)

```

```python
#Evaluating  the best model
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best Parameters:", best_params)
print("Best Cross-Validation Score:", best_score)
```

```python
# Evaluate the best model on the test set
y_pred = best_model.predict(X_test)
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

# 4. Evaluation

```python

# Evaluate the model using various metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate accuracy, precision, recall, and F1-score for the default model
default_model_accuracy = accuracy_score(y_test, model.predict(X_test).argmax(axis=1))
default_model_precision = precision_score(y_test, model.predict(X_test).argmax(axis=1), average='weighted')
default_model_recall = recall_score(y_test, model.predict(X_test).argmax(axis=1), average='weighted')
default_model_f1_score = f1_score(y_test, model.predict(X_test).argmax(axis=1), average='weighted')

# Calculate accuracy, precision, recall, and F1-score for the tuned model
tuned_model_accuracy = accuracy_score(y_test, best_model.predict(X_test))
tuned_model_precision = precision_score(y_test, best_model.predict(X_test), average='weighted')
tuned_model_recall = recall_score(y_test, best_model.predict(X_test), average='weighted')
tuned_model_f1_score = f1_score(y_test, best_model.predict(X_test), average='weighted')

# Print the results
print("Default Model Metrics:")
print("- Accuracy:", default_model_accuracy)
print("- Precision:", default_model_precision)
print("- Recall:", default_model_recall)
print("- F1-Score:", default_model_f1_score)

print("Tuned Model Metrics:")
print("- Accuracy:", tuned_model_accuracy)
print("- Precision:", tuned_model_precision)
print("- Recall:", tuned_model_recall)
print("- F1-Score:", tuned_model_f1_score)




```
