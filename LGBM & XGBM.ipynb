# Data Preprocessing:

```python
# Impute missing values in Age with the median
df1['Age'].fillna(df1['Age'].median(), inplace=True)
df2['Age'].fillna(df2['Age'].median(), inplace=True)

# Impute missing values in Fare with the mean
df1['Fare'].fillna(df1['Fare'].mean(), inplace=True)
df2['Fare'].fillna(df2['Fare'].mean(), inplace=True)

# Impute missing values in Embarked with the most frequent value
df1['Embarked'].fillna(df1['Embarked'].mode()[0], inplace=True)
df2['Embarked'].fillna(df2['Embarked'].mode()[0], inplace=True)
```

```python
# Encode categorical variables.

#data transformation for train data
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
for column in df1.columns:
    df1[column] = LE.fit_transform(df1[column])
df1.head()
```

```python
#data transformation for test data
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
for column in df2.columns:
    df2[column] = LE.fit_transform(df2[column])
df2.head()
```

# Building Predictive Models:

```python
# define x and y

X=df1.drop('Survived', axis=1)
Y=df1['Survived']
```

```python
# 1.Split the preprocessed dataset into training and testing sets.

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

from sklearn.linear_model import LogisticRegression
Logreg = LogisticRegression()

# fit the model
Logreg.fit(X_train, Y_train)
# predictions
y_pred = Logreg.predict(X_test)
y_pred

# Train the model
Logreg.fit(X_train, Y_train)

# 2.Choose appropriate evaluation metrics (e.g., accuracy, precision, recall, F1-score) for model evaluation.


y_pred = Logreg.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(Y_test, y_pred)
precision = precision_score(Y_test, y_pred)
recall = recall_score(Y_test, y_pred)
f1 = f1_score(Y_test, y_pred)
roc_auc = roc_auc_score(Y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc)
```

```python
# 3. Build predictive models using LightGBM and XGBoost algorithms.
# 4. Train the models on the training set and evaluate their performance on the testing set.

# Import necessary libraries
#!pip install lightgbm
#!pip install xgboost
import lightgbm as lgb
import xgboost as xgb



# Train a LightGBM model
lgb_model = lgb.LGBMClassifier()
lgb_model.fit(X_train, Y_train)

# Train an XGBoost model
xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train, Y_train)

# Make predictions using both models
lgb_pred = lgb_model.predict(X_test)
xgb_pred = xgb_model.predict(X_test)
```

```python

# Calculate accuracy, precision, recall, and F1-score for both models
print("LightGBM model:")
lgb_accuracy = accuracy_score(Y_test, lgb_pred)
lgb_precision = precision_score(Y_test, lgb_pred)
lgb_recall = recall_score(Y_test, lgb_pred)
lgb_f1 = f1_score(Y_test,lgb_pred)
print("LightGBM Accuracy:", lgb_accuracy)
print("LightGBM precision:", lgb_precision)
print("LightGBM recall:", lgb_recall)
print("LightGBM f1 score:", lgb_f1)

print("#===================================================")

xgb_accuracy = accuracy_score(Y_test, xgb_pred)
xgb_precision = precision_score(Y_test, xgb_pred)
xgb_recall = recall_score(Y_test, xgb_pred)
xgb_f1 = f1_score(Y_test, xgb_pred)
print("XGBoost model:")
print("XGBoost Accuracy:", xgb_accuracy)
print("XGBoost precision:", xgb_precision)
print("XGBoost recall:", xgb_recall)
print("XGBoost f1 score:", xgb_f1)




```

```python
# 5. Use techniques like cross-validation and hyperparameter tuning to optimize model performance.

from sklearn.model_selection import cross_val_score

# Perform cross-validation on the Logistic Regression model
cv_logreg = cross_val_score(Logreg, X, Y, cv=10)
print("Logistic Regression Cross-Validation Scores:", cv_logreg)
print("#===================================================")

# cross-validation on the LightGBM model
cv_lgbm = cross_val_score(lgb_model, X, Y, cv=10)
print("LightGBM Cross-Validation Scores:", cv_lgbm)
print("#===================================================")

# cross-validation on the XGBoost model
cv_xgb = cross_val_score(xgb_model, X, Y, cv=10)
print("XGBoost model Cross-Validation Scores:", cv_xgb)
print("#===================================================")


# Print the average cross-validation score
print("Average Cross-Validation Score (Logistic Regression):", cv_logreg.mean())
print("Average Cross-Validation Score (LightGBM):", cv_lgbm.mean())
print("Average Cross-Validation Score (XGBoost):", cv_xgb.mean())
```

```python
# Define the grid of hyperparameters 
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.1, 0.2, 0.3],
    'max_depth': [None, 5, 10],
}

# Perform grid search using the LightGBM model
grid_search_lgbm = GridSearchCV(lgb_model, param_grid, cv=5)
grid_search_lgbm.fit(X_train, Y_train)

# Perform grid search using the XGBoost model
grid_search_xgb = GridSearchCV(xgb_model, param_grid, cv=5)
grid_search_xgb.fit(X_train, Y_train)

# Print the best hyperparameters
print("Best Hyperparameters (LightGBM):", grid_search_lgbm.best_params_)

print("Best Hyperparameters (XGBoost):", grid_search_xgb.best_params_)

```

```python
# 1. Compare the performance metrics of LightGBM and XGBoost models

# Create a table to compare the performance metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
lgbm_metrics = [lgb_accuracy, lgb_precision, lgb_recall,lgb_f1]
xgb_metrics = [xgb_accuracy, xgb_precision, xgb_recall,xgb_f1]

comparison_table = pd.DataFrame({
    'Metric': metrics,
    'LightGBM': lgbm_metrics,
    'XGBoost': xgb_metrics
})

print(comparison_table)

# 2. Visualize and interpret the results

# Create a bar chart to compare the performance metrics
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(metrics, lgbm_metrics, label='LightGBM', alpha=0.8)
plt.bar(metrics, xgb_metrics, label='XGBoost', alpha=0.8)

plt.xlabel('Metric')
plt.ylabel('Score')
plt.title('Performance Comparison of LightGBM and XGBoost Models')
plt.legend()
plt.show()

```
