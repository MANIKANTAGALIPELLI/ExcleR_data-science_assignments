```python
import pandas as pd

# Load the data
df=pd.read_csv('D:\\Data science\\assignments\\sales_data_with_discounts.csv')
df
df.head
df.info()
```

```python
#There are nulls in the data set so,we should replace them with 0
df.fillna(0,inplace=True)
df
```

```python
# Identify numerical columns
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns
# Print the numerical columns
print(numerical_columns)
```

```python
#Calculate the mean, median, mode, and standard deviation for numerical_columns

# Calculate the mean for each numerical column

mean=df[numerical_columns].mean().round(2)
print(mean)
```

```python
# Calculate the median for each numerical column

median=df[numerical_columns].median().round(2)
print(median)
```

```python
# Calculate the mode for each numerical column

mode=df[numerical_columns].mode().round(2)
print(mode)
```

```python
# Calculate the standard deviation for each numerical column

standarddeviation =df[numerical_columns].std().round(2)
print(standarddeviation)
```

```python
#By interpreting these statistics, we can gain a better understanding of the central tendencies and dispersion of the data within each numerical column of the "sales_data_with_discounts.csv" dataset.
#Plot histograms for each numerical column.

import matplotlib.pyplot as plt

# Iterate through each numerical column and plot its histogram
df['Volume'].hist()  
plt.title("Histograms of Volume ")
plt.show()
df['Avg Price'].hist() 
plt.title("Histograms of Avg Price ")
plt.show()
df['Total Sales Value'].hist()
plt.title("Histograms of Total Sales Value ")
plt.show()
df['Discount Rate (%)'].hist()
plt.title("Histograms of Discount Rate (%) ")
plt.show()
df['Discount Amount'].hist()
plt.title("Histograms of Discount Amount ")
plt.show()
df['Net Sales Value'].hist()
plt.title("Histograms of Net Sales Value ")
plt.show()
```

```python
#skewness of Volume is positive
#skewness of Avg Price is  positive
#skewness of Total Sales Value is positive
#skewness of Discount Rate (%) is negative
#skewness of Discount Amount is positive
#skewness of Net Sales Value is positive
```

```python
#boxplots for numerical variables 
plt.figure(figsize=(10,8))
df.boxplot(column=['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value'],vert=False)
plt.title('Boxplots of Numerical Variables')
plt.xlabel('Variable')
plt.ylabel('Value')
plt.show()
```

```python
#Identify categorical columns in the dataset.

categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)
```

```python
#Create bar charts to visualize the frequency or count of each category

for col in categorical_columns:
  values = df[col].value_counts()
# Create a bar chart for the current column
  plt.figure(figsize=(10, 6))
  plt.bar(values.index, values.values)
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.title('Bar Chart for {col}')
  plt.show()

```

```python
# Standardization of Numerical Variables
# Standardization is a process that remove the units from the variable and put the data in around -3 to 3

import numpy as np

# Standardize numerical columns
zscore_df=(df[numerical_columns] - mean) / standarddeviation
# Print the standardized data
print(zscore_df.head())

# Before standardization
print("Before standardization:")
print(df[numerical_columns].describe())

# After standardization
print("\nAfter standardization:")
print(zscore_df.describe())
```

```python
# converting categorical data into dummy variables 
#one-hot encoding to the categorical columns
encoded_data = pd.get_dummies(df[categorical_columns], drop_first=True)
print(encoded_data.head())

# Display a portion of the transformed dataset
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded_data = encoded_data.apply(le.fit_transform)
print(encoded_data.head())
```
