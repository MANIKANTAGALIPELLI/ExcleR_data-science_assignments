```python
# Impute missing values for numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Impute missing values for categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
# Apply scaling techniques to numerical features:
# â€¢	Standard Scaling
# â€¢	Min-Max Scaling

from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Standard Scaling
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df[numerical_columns])
scaled_df = pd.DataFrame(scaled_df, columns=numerical_columns)

# Min-Max Scaling
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df[numerical_columns])
scaled_df = pd.DataFrame(scaled_df, columns=numerical_columns)
```

# 2. Encoding Techniques:

```python

# One-Hot Encoding for categorical variables with less than 5 categories
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
  if df[col].nunique() < 5:
    df_dummies = pd.get_dummies(df[col], prefix=col)
    df = pd.concat([df, df_dummies], axis=1)
    df.drop(col, axis=1, inplace=True)
df.head()
```

```python
# Label Encoding for categorical variables with more than 5 categories
for col in categorical_columns:
  if df[col].nunique() > 5:
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
# If Label Encoding gets error,once again run One-Hot Encoding and then label encoding
```

```python
# Pros and cons of One-Hot Encoding:

# Pros:

# Easy to interpret
# Creates new features for each category

# Cons:

# Can increase the dimensionality of the data
# Can be computationally expensive
#========================================================================================================
# Pros and cons of Label Encoding:

# Pros:

# Simple to implement
# Does not increase the dimensionality of the data

# Cons:

# Can be misleading if the categories are not ordered
# Can lead to overfitting if the categories are imbalanced
```

# 3. Feature Engineering:

```python
# Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices.

# Feature 1: Ratio of education years to work experience years

df['education_work_ratio'] = df['education_num'] / df['hours_per_week']
print("Feature 1: Ratio of education years to work experience years:",df['education_work_ratio'])

# Feature 2: Product of education years and occupation code

df['education_occupation_product'] = df['education_num'] * df['occupation']
print("Feature 2: Product of education years and occupation code:",df['education_occupation_product'])

# Apply log transformation to positively skewed numerical feature

import numpy as np
df['log_capital_gain'] = np.log(df['capital_gain'] + 1)

print("Apply log transformation to positively skewed numerical feature",df['log_capital_gain'])

# Justify the choice
print("Justify the choice:")
print("Log transformation: The capital-gain feature is positively skewed which can affect the performance of some machine learning models.")

```

# 4. Feature Selection:

```python
# Outlier Detection and Removal using Isolation Forest

from sklearn.ensemble import IsolationForest

# Create an instance of the IsolationForest model
model = IsolationForest(contamination='auto')

# Fit the model to the data
model.fit(df[numerical_columns])

# Get the outlier scores
outlier_scores = model.decision_function(df[numerical_columns])

# Get the outlier labels
outlier_labels = model.predict(df[numerical_columns])

# Remove outliers
df_without_outliers = df[outlier_labels != -1]

df_without_outliers

# Discussion:
# Outliers can affect model performance in several ways:
```
